```python
#seção imports
import findspark
import os
import pyspark.sql.functions as R
```

```python
from pyspark.sql import SparkSession
```

```python
# Este seção cria o APPNAME e configura o formato de saida da função show()
spark = ( 
    SparkSession.builder.appName('ralves_sess_2')
    .config ("spark.sql.repl.eagerEval.enabled",True)
    .getOrCreate()
)
```

```python
#file shortcut 
trnfile2=r'C:/sources/LOGINS.parquet'

#create dataframe
df=spark.read.parquet(trnfile2)
```

```python
df
```

```python
#usando o help do pyspark para mostrar as opções da biblioteca FUNCTION
help(R)
```

```python
df.withColumn('owner',R.lit('ralves'))
```

```python
df.withColumn('nome_estado',R.when(df.estado=='MG','Minas Gerais'))
```

```python
df1=df.select('cpf', 'profissao', 'telefone','email')
```

```python
df1.show(20,False)
```

```python
df1.withColumn('nome_usuario',R.split(df1.email,'@'))
```

```python
df1.orderBy('cpf')
```

```python
dfsummary=df.groupBy('estado','cor_favorita').agg(R.count('*')).orderBy('estado','cor_favorita')
```

```python
dfsummary.show()
```

```python
dfsummary.groupBy('estado').agg(R.sum('count(1)')).orderBy('estado')
```
